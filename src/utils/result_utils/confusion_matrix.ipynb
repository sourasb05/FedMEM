```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os


def confusion_matrix(folder_path, save_path):

    # List all CSV files in the folder
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

    # Initialize the sum of confusion matrices with zeros
    # Assuming the shape of confusion matrices is known, e.g., 2x2. Adjust it according to your actual matrices.
    #  You might need to adjust this after reading the first file if the size could vary.
    # for file_name in csv_files:
    #    file_path = os.path.join(folder_path, file_name)
    #    cm = pd.read_csv(file_path).values  # Reading the confusion matrix into a numpy array


    # Set up the subplot grid
    nrows, ncols = 8, 5  # Adjust based on your needs
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(24, 20))
    fig.tight_layout(pad=6.0)

    # Iterate over each file and plot the confusion matrix
    for i, file_name in enumerate(csv_files):
        # Calculate row and column index for the subplot
        row = i // ncols
        col = i % ncols

        # Read the confusion matrix from the CSV file
        cm_path = os.path.join(folder_path, file_name)
        cm = pd.read_csv(cm_path).to_numpy()

        # Plot the confusion matrix
        ax = axes[row, col]
        cax = ax.matshow(cm, cmap=plt.cm.Blues)
        fig.colorbar(cax, ax=ax)
    
        ax.title.set_text(f'{file_name[3:-4]}')

        # Optionally, add numbers to the confusion matrix
        for (j, k), val in np.ndenumerate(cm):
            ax.text(k, j, f'{val}', ha='center', va='center', color='red')

    # Adjust subplot parameters or save figure
    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.50, wspace=0.35)
    plt.suptitle('Confusion Matrices for Each User', fontsize=16)
    plt.savefig(save_path + '/confusion_matrices_subplot.png')
    plt.show()

        
    
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os


def save_confusion_matrix(folder_path, save_path):

    # List all CSV files in the folder
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

    # Ensure the save path exists
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    # Iterate over each file and save the confusion matrix
    for file_name in csv_files:
        # Read the confusion matrix from the CSV file
        cm_path = os.path.join(folder_path, file_name)
        cm = pd.read_csv(cm_path).to_numpy()  # Added header=None for CSVs without header

        # Create a figure and axis for the plot
        fig, ax = plt.subplots(figsize=(6, 5))
        
        # Use 'imshow' to plot the confusion matrix
        cax = ax.imshow(cm, cmap=plt.cm.Blues)
        fig.colorbar(cax, ax=ax)

        # Add title
        # ax.set_title(f'{file_name[:-4]}', pad=20)
        
        ax.tick_params(axis='both', which='major', labelsize=20)

        # Optionally, add numbers to the confusion matrix
        for (i, j), val in np.ndenumerate(cm):
            ax.text(j, i, f'{val}', ha='center', va='center', color='red', fontsize=20)

        # Save the figure
        plt.savefig(os.path.join(save_path + '/plot', f'{file_name[3:-4]}.png'))
        plt.close(fig)  # Ensure the figure is closed after saving

        
    
    
```

```python
# Specify the folder where your CSV files are stored
folder_path = '/proj/sourasb-220503/FedMEM/results/confusion_matrix/ResNet50TL/Fedmem/data_silo_100/target_10/dynamic/40.0/0'
save_path = '/proj/sourasb-220503/FedMEM/results/confusion_matrix/ResNet50TL/Fedmem/data_silo_100/target_10/dynamic/40.0/0'
save_confusion_matrix(folder_path, save_path)
```
