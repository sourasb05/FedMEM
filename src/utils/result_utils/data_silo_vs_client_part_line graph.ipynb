```python
import pandas as pd
import matplotlib.pyplot as plt


# Create a dictionary with data
data_1 = {
    '20': [19.70, 23.86, 26.06, 26.58, 27.39],
    '40': [19.57, 27.45, 27.48, 28.13, 28.82],
    '60': [19.18, 25.81, 27.46, 28.21, 28.87],
    '80': [19.39, 28.13, 28.25, 28.83, 29.10],
    '100': [20.17, 28.41, 28.32, 28.71, 28.69]
    
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data_1)

# X-axis labels (assumed as indexes for the F1-Scores in each column)
x_labels = [20, 40, 60, 80, 100]
y_labels = [15, 20, 25, 30, 35, 40]

# Create a plot
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green', 'purple', 'black']

# Plot each column as a separate line
for i, column in enumerate(df.columns):
    plt.plot(x_labels, df[column], marker='o', label=f'{column}', color=colors[i])

# Adding labels and title
plt.xlabel('% of Lifeloggers Participating', fontsize=24)
plt.ylabel('F1-Score(%)', fontsize=24)
# plt.title('F1-Score vs. Data P', fontsize=20)
# Set x-ticks to ensure they appear at intervals of 20
plt.xticks(x_labels, fontsize=18)
plt.yticks(y_labels, fontsize=18)
# Show legend
plt.legend(title="% of data per lifelogger",title_fontsize=24,ncol=3, fontsize=20)

# Show grid
plt.grid(True)
plt.tight_layout()

plt.savefig('/proj/sourasb-220503/FedMEM/results/ResNet50TL/Fedmem/data_vs_client_figute/apriori.png')
# Show the plot
plt.show()
```
