```python
import os

# Get list of file paths for the score files
score_files = [f"/proj/sourasb-220503/FedMEM/dataset/labels/score_{score}.csv" for score in range(1, 10)]

# Function to read and combine files in pairs
def combine_files_in_pairs(files):
    combined_data = []
    for i in range(0, len(files), 2):
        # Read the data from two files
        data1 = pd.read_csv(files[i])
        if i+1 < len(files):
            data2 = pd.read_csv(files[i+1])
        else:
            # If there's an odd number of files, just use the last file alone
            data2 = pd.DataFrame(columns=['name', 'score'])
        
        # Combine the two DataFrames
        combined = pd.concat([data1, data2], ignore_index=True)
        combined_data.append(combined)
    return combined_data

# Combine the files in pairs
combined_data_pairs = combine_files_in_pairs(score_files)

# Save the combined data to new files
combined_filenames = []
for i, data in enumerate(combined_data_pairs):
    filename = f"/proj/sourasb-220503/FedMEM/dataset/n_iid_2/combined_data_{i+1}.csv"
    data.to_csv(filename, index=False)
    combined_filenames.append(filename)

combined_filenames
```
