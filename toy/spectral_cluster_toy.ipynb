```python
from sklearn.metrics.pairwise import rbf_kernel

# Create the similarity matrix using a Gaussian kernel
gamma = 15  # This parameter can be tuned
similarity_matrix = rbf_kernel(X, gamma=gamma)
print(similarity_matrix)
```

```python
# Compute the Degree Matrix
degree_matrix = np.diag(similarity_matrix.sum(axis=1))
print(degree_matrix)
# Compute the Laplacian Matrix
laplacian_matrix = degree_matrix - similarity_matrix
print(laplacian_matrix)
```

```python
eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)

print(eigenvalues)
print(eigenvectors)
```

```python
# Select the first two eigenvectors for 2 clusters
n_clusters = 5
X_transformed = eigenvectors[:, :n_clusters]

print(X_transformed)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X_transformed)
labels = kmeans.labels_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k')
plt.title("Clusters Identified by Spectral Clustering")
plt.show()
```
