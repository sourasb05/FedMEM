```python
import pandas as pd
import os
import csv
```

1. change .jpg to .pt

```python
# Path to the original file
original_file_path = '/proj/sourasb-220503/FedMEM/dataset/train_1.txt'

# Path to the new file with modified filenames
new_file_path = '/proj/sourasb-220503/FedMEM/dataset/tensor_train_1.txt'

# Open the original file to read and the new file to write
with open(original_file_path, 'r') as original_file, open(new_file_path, 'w') as new_file:
    # Iterate through each line in the original file
    for line in original_file:
        # Split the line into filename and label
        parts = line.strip().split(' ')
        if len(parts) == 2:
            filename, label = parts
            # Replace the .jpg extension with .pt
            new_filename = filename.replace('.jpg', '.pt')
            # Write the modified filename and label to the new file
            new_file.write(f'{new_filename} {label}\n')

print(f'Modified filenames have been saved to {new_file_path}.')
```

2. Convert txt to csv

```python

def create_csv_file(input_file, output_file):
    """Creates a CSV file from a text file that contains two columns separated by space.

    Args:
        input_file: The path to the input text file.
        output_file: The path to the output CSV file.
    """

    with open(input_file, "r") as f_in, open(output_file, "w", newline="") as f_out:
        writer = csv.writer(f_out)
        for line in f_in:
            values = line.split()
            writer.writerow(values)

if __name__ == "__main__":
    input_file = "/proj/sourasb-220503/FedMEM/dataset/tensor_train_val_1.txt"
    output_file = "/proj/sourasb-220503/FedMEM/dataset/tensor_train_val_1.csv"

    create_csv_file(input_file, output_file)
```

To check if the first column of a CSV file contains entries that correspond to .pt (PyTorch model files, for example) filenames in a specific folder, you can follow these steps using Python. This involves reading the CSV file, extracting the first column's contents, and then comparing these to the list of .pt files in your target directory.

```python
# Replace these paths with your actual paths
csv_file_path = '/proj/sourasb-220503/FedMEM/dataset/cleaned_tensor_train_val_1.csv'
folder_path = '/proj/sourasb-220503/FedMEM/dataset/r3_mem_ResNet50_features'

# Read the CSV file and extract the first column
df = pd.read_csv(csv_file_path)
first_column = df.iloc[:, 0].tolist() # Assuming the first column contains the filenames

# List all .pt files in the specified folder
pt_files = [f for f in os.listdir(folder_path) if f.endswith('.pt')]
print(pt_files)
print(first_column)

set1 = set(pt_files)
set2 = set(first_column)

# Find common entries
common_entries = set1.intersection(set2)
print(f"Common entries ({len(common_entries)}): {common_entries}")


# Find entries unique to list1
unique_to_list1 = set1 - set2
print(f"Entries unique to list1: {unique_to_list1}")

# Find entries unique to list2
unique_to_list2 = set2 - set1
print(f"Entries unique to list2: {unique_to_list2}")

# Alternatively, find all differences (symmetric difference)
all_differences = set1.symmetric_difference(set2)
print(f"All differences: {all_differences}")
```
